```python
import os
from jamo import h2j, j2hcj

dataset_path = '/root/paper_project/ML/sample_data'

def get_labels(root_folder_path):
    label_names = []
    for label in os.listdir(root_folder_path):
        label_names.append(label)
    return label_names

labels = get_labels(dataset_path)

chosung_list = ['ㄱ', 'ㄲ', 'ㄴ', 'ㄷ', 'ㄸ', 'ㄹ', 'ㅁ', 'ㅂ', 'ㅃ', 'ㅅ', 'ㅆ', 'ㅇ', 'ㅈ', 'ㅉ', 'ㅊ', 'ㅋ', 'ㅌ', 'ㅍ', 'ㅎ']
jungsung_list = ['ㅏ', 'ㅐ', 'ㅑ', 'ㅒ', 'ㅓ', 'ㅔ', 'ㅕ', 'ㅖ', 'ㅗ', 'ㅘ', 'ㅙ', 'ㅚ', 'ㅛ', 'ㅜ', 'ㅝ', 'ㅞ', 'ㅟ', 'ㅠ', 'ㅡ', 'ㅢ', 'ㅣ']
jonsung_list = [' ', 'ㄱ', 'ㄲ', 'ㄳ', 'ㄴ', 'ㄵ', 'ㄶ', 'ㄷ', 'ㄹ', 'ㄺ', 'ㄻ', 'ㄼ', 'ㄽ', 'ㄾ', 'ㄿ', 'ㅀ', 'ㅁ', 'ㅂ', 'ㅄ', 'ㅅ', 'ㅆ', 'ㅇ', 'ㅈ', 'ㅊ', 'ㅋ', 'ㅌ', 'ㅍ', 'ㅎ']

chosung_dict = {chosung: 0 for idx, chosung in enumerate(chosung_list)}
jungsung_dict = {jungsung: 0 for idx, jungsung in enumerate(jungsung_list)}
jonsung_dict = {jonsung: 0 for idx, jonsung in enumerate(jonsung_list)}

for label in labels:
    if len(j2hcj(h2j(label))) == 2:
        first, middle = j2hcj(h2j(label))
        chosung_dict[first] += 1
        jungsung_dict[middle] += 1
    else:
        first, middle, end = j2hcj(h2j(label))
        chosung_dict[first] += 1
        jungsung_dict[middle] += 1
        jonsung_dict[end] += 1


print(chosung_dict)
print(jungsung_dict)
print(jonsung_dict)
```
