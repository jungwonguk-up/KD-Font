```python
import os
from jamo import h2j, j2hcj

dataset_path = '/root/paper_project/ML/sample_data'

text_path = '/root/paper_project/Tools/DataImbalance/sampling_letter.txt'

sample_420 = []

with open(text_path, 'r', encoding='utf-8') as file:
    for line in file:
        sample_420.append(line.strip())

def get_labels(root_folder_path):
    label_names = []
    for label in os.listdir(root_folder_path):
        label_names.append(label)
    return label_names

labels = get_labels(dataset_path)

# labels = sample_420

chosung_list = ['ㄱ', 'ㄲ', 'ㄴ', 'ㄷ', 'ㄸ', 'ㄹ', 'ㅁ', 'ㅂ', 'ㅃ', 'ㅅ', 'ㅆ', 'ㅇ', 'ㅈ', 'ㅉ', 'ㅊ', 'ㅋ', 'ㅌ', 'ㅍ', 'ㅎ']
jungsung_list = ['ㅏ', 'ㅐ', 'ㅑ', 'ㅒ', 'ㅓ', 'ㅔ', 'ㅕ', 'ㅖ', 'ㅗ', 'ㅘ', 'ㅙ', 'ㅚ', 'ㅛ', 'ㅜ', 'ㅝ', 'ㅞ', 'ㅟ', 'ㅠ', 'ㅡ', 'ㅢ', 'ㅣ']
jonsung_list = [' ', 'ㄱ', 'ㄲ', 'ㄳ', 'ㄴ', 'ㄵ', 'ㄶ', 'ㄷ', 'ㄹ', 'ㄺ', 'ㄻ', 'ㄼ', 'ㄽ', 'ㄾ', 'ㄿ', 'ㅀ', 'ㅁ', 'ㅂ', 'ㅄ', 'ㅅ', 'ㅆ', 'ㅇ', 'ㅈ', 'ㅊ', 'ㅋ', 'ㅌ', 'ㅍ', 'ㅎ']

chosung_dict = {chosung: 0 for idx, chosung in enumerate(chosung_list)}
jungsung_dict = {jungsung: 0 for idx, jungsung in enumerate(jungsung_list)}
jonsung_dict = {jonsung: 0 for idx, jonsung in enumerate(jonsung_list)}

for label in labels:
    if len(j2hcj(h2j(label))) == 2:
        first, middle = j2hcj(h2j(label))
        chosung_dict[first] += 1
        jungsung_dict[middle] += 1
    else:
        first, middle, end = j2hcj(h2j(label))
        chosung_dict[first] += 1
        jungsung_dict[middle] += 1
        jonsung_dict[end] += 1


print(chosung_dict)
print(jungsung_dict)
print(jonsung_dict)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 프레임 생성
df_chosung = pd.DataFrame(list(chosung_dict.items()), columns=['Chosung', 'Count'])
df_jungsung = pd.DataFrame(list(jungsung_dict.items()), columns=['Jungsung', 'Count'])
df_jongsung = pd.DataFrame(list(jonsung_dict.items()), columns=['Jongsung', 'Count'])

# 색상 리스트
colors_chosung = ['lightgreen'] * len(df_chosung)
colors_jungsung = ['lightgreen'] * len(df_jungsung)
colors_jongsung = ['lightgreen'] * len(df_jongsung)

# 최솟값이 0이 아닌 컬럼 중 가장 작은 값을 찾습니다.
min_count_chosung = min(df_chosung[df_chosung['Count'] > 0]['Count'])
min_count_jungsung = min(df_jungsung[df_jungsung['Count'] > 0]['Count'])
min_count_jongsung = min(df_jongsung[df_jongsung['Count'] > 0]['Count'])

# 최댓값을 찾습니다.
max_count_chosung = max(df_chosung['Count'])
max_count_jungsung = max(df_jungsung['Count'])
max_count_jongsung = max(df_jongsung['Count'])

# 최솟값이 0이 아닌 컬럼의 인덱스를 찾습니다.
min_index_chosung = df_chosung[df_chosung['Count'] == min_count_chosung].index[0]
min_index_jungsung = df_jungsung[df_jungsung['Count'] == min_count_jungsung].index[0]
min_index_jongsung = df_jongsung[df_jongsung['Count'] == min_count_jongsung].index[0]

# 최댓값의 인덱스를 찾습니다.
max_index_chosung = df_chosung[df_chosung['Count'] == max_count_chosung].index[0]
max_index_jungsung = df_jungsung[df_jungsung['Count'] == max_count_jungsung].index[0]
max_index_jongsung = df_jongsung[df_jongsung['Count'] == max_count_jongsung].index[0]

# 해당 인덱스에 다른 색상을 적용 (초성)
colors_chosung[min_index_chosung] = 'lightcoral'
# 해당 인덱스에 다른 색상을 적용 (중성)
colors_jungsung[min_index_jungsung] = 'lightcoral'
# 해당 인덱스에 다른 색상을 적용 (종성)
colors_jongsung[min_index_jongsung] = 'lightcoral'

# 가장 많은 것에 다른 색상을 적용 (초성)
colors_chosung[max_index_chosung] = 'lightskyblue'
# 가장 많은 것에 다른 색상을 적용 (중성)
colors_jungsung[max_index_jungsung] = 'lightskyblue'
# 가장 많은 것에 다른 색상을 적용 (종성)
colors_jongsung[max_index_jongsung] = 'lightskyblue'

# 막대 그래프 그리기
plt.figure(figsize=(15, 5))
plt.subplot(131)
plt.bar(df_chosung['Chosung'], df_chosung['Count'], color=colors_chosung)
plt.xlabel('Chosung')
plt.ylabel('Count')
plt.title('Chosung Counts')

plt.subplot(132)
plt.bar(df_jungsung['Jungsung'], df_jungsung['Count'], color=colors_jungsung)
plt.xlabel('Jungsung')
plt.ylabel('Count')
plt.title('Jungsung Counts')

plt.subplot(133)
plt.bar(df_jongsung['Jongsung'], df_jongsung['Count'], color=colors_jongsung)
plt.xlabel('Jongsung')
plt.ylabel('Count')
plt.title('Jongsung Counts')

plt.tight_layout()
plt.show()
```
